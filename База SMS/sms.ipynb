```python
# устанавим необходимые библиотеки, в т.ч. SQLite3
import pandas as pd
import numpy as np
import sqlite3
print(sqlite3.sqlite_version) # 3.40.1
```

```python
# создадим отдельно таблицы и заполним их произвольными значениями,загрузим
accounts = pd.read_excel('/accounts.xlsx')
blacklist = pd.read_csv('/blacklist.csv')
card_status = pd.read_excel('/card_status.xlsx')
cards = pd.read_excel('/cards.xlsx')
catalog = pd.read_excel('/catalog.xlsx')
tel_catalog = pd.read_excel('/tel_catalog.xlsx')
comms = pd.read_excel('/comms.xlsx')
```

```python
# создадим соединение с базой данных SQLite, представленной файлом 'sms'
con = sqlite3.connect('sms')
```

```python
# создадим объект курсора, который позволит выполнить SQL-запросы к базе данных, связанной с соединением 'con'
cur = con.cursor()
```

```python
accounts.to_sql('accounts',con,index=False,if_exists='replace')
blacklist.to_sql('blacklist',con,index=False,if_exists='replace')
card_status.to_sql('card_status',con,index=False,if_exists='replace')
catalog.to_sql('catalog',con,index=False,if_exists='replace')
cards.to_sql('cards',con,index=False,if_exists='replace')
tel_catalog.to_sql('tel_catalog',con,index=False,if_exists='replace')
comms.to_sql('comms',con,index=False,if_exists='replace')
```

Напишем функцию,которая облегчает выполнение SQL-запросов к базе данных и возвращает результат в формате DataFrame, что делает его более удобным для анализа и обработки данных с использованием библиотеки pandas.

```python
def select(sql):
  return pd.read_sql(sql,con)
```

```python
sql = '''with temp_table(client_id, hash_id, rest_rub, tarif_name, tel_num, rank_cart, date_activated, neg)
         AS (
        select t.client_id, t.hash_id, u.rest_rub, r.tarif_name, i.tel_num,
        RANK() OVER(PARTITION BY t.client_id ORDER BY u.rest_rub desc) as rank_cart,
        t.date_activated, i.tel_num % 2
        from cards t

        left join tel_catalog i ON t.client_id=i.client_id
        left join accounts u ON t.acc_num=u.acc_num
        left join card_status y ON t.hash_id=y.hash_id
        left join catalog r ON t.tarif_id=r.tarif_id

        where t.client_id NOT IN (select * from blacklist x)
        and r.tarif_name = 'амурский тигр'
        and t.date_activated BETWEEN '2016-04-01' AND '2016-06-30'
        and u.rest_rub > 1000
        and DATE(u.report_date) = '2016-07-01'
        and y.status = 'open')


        select tt.client_id, tt.hash_id, tt.rest_rub, tt.tarif_name, tt.tel_num,
        case when tt.tel_num % 2 = 1
        then strftime('%m', tt.date_activated) ||', ' || (tt.rest_rub*1.0/1000) || ' тыс.руб, ' 
                 || substring(tt.tel_num, 2, 4)
        else 'Здравствуйте '|| tt.client_id || ', ' ||strftime('%m', tt.date_activated) ||', ' 
                 || (tt.rest_rub*1.0/1000) || ' тыс.руб, '
        || substring(tt.tel_num, 2, 4) end as sms_text
        from temp_table tt
        where tt.rank_cart = 1


        '''
select(sql)
```

1. Создали временную таблицу `temp_table` с несколькими столбцами: `client_id`, `hash_id`, `rest_rub`, `tarif_name`, `tel_num`, `rank_cart`, `date_activated` и `neg`.   
  Столбец `neg` в таблице `temp_table` создается внутри `SELECT` подзапроса и используется для вычисления значения столбца `tel_num % 2`.
  Выражение `tel_num % 2` вычисляет остаток от деления значения столбца `tel_num` на 2. Если это значение равно 1, то в столбец `neg` будет записано значение 1; в противное случае, если значение равно 0, в столбец `neg` будет записано значение 0.
  Столбец `neg` будет использоваться в дальнейшем, в выражении `case when p.tel_num % 2 = 1 ... else ... end`. Если значение столбца  `tel_num % 2` равно 1, то будет выполняться условие `when p.tel_num % 2 = 1`, а если значение равно 0, то будет выполняться условие `else`.   
2. В `SELECT`-подзапросе мы заполнили `temp_table`, выбирая значение каждого столбца из различных таблиц с использованием операторов `LEFT JOIN`.
3. Использовали функцию `RANK() OVER(PARTITION BY t.client_id ORDER BY u.rest_rub desc) as rank_cart` для присвоения ранга каждой записи в таблице `temp_table` на основе значения столбца `rest_rub` по возрастанию, сгруппированного по `client_id` (т.к.по условию нам надо выбрать карту с максимальным остатком).
4. В `SELECT`-подзапросе выбираются только те строки из `temp_table`, у которых `rank_cart = 1`.
5. В соответствии с условием возвращаются следующие поля: значения столбцов `client_id`, `hash_id`, `rest_rub`, `tarif_name`, `tel_num`, а также текстовое значение столбца `sms_text`.
6. Поле `sms_text` в основной выборке формируется следующим образом:
  Здесь используется функция `strftime('%m', p.date_activated)`, которая извлекает месяц из даты `date_activated` в формате MM (например, 01 для января, 02 для февраля и т.д.). Затем к полученному значению добавляется строка `', '`.

  Далее используется арифметическое выражение `(p.rest_rub*1.0/1000)`, которое делит значение `rest_rub` на 1000 и преобразует его в число с плавающей запятой. Результат этого выражения представляется в виде строки и добавляется к уже составленной строке `тыс.руб, `.

  Затем используется функция `substring(p.tel_num, 2, 4)`, которая извлекает подстроку из поля `tel_num`, начиная с позиции 2 и длиной 4 символа. Результат подстроки также добавляется к формирующейся строке `тыс.руб, `.

  Наконец, для формирования полного `sms_text` используется условное выражение `case when p.tel_num % 2 = 1 ... else ... end`. Если последняя цифра номера телефона нечетная (`p.tel_num % 2 = 1`), то добавляется строка `'месяц, остаток средств в тыс.руб, 4 последние цифры номера'`. Если последняя цифра номера телефона четная, то добавляется строка `'Здравствуйте идентификатор клиента, месяц, остаток средств в тыс.руб, со 2 по 4 цифры номера телефона'`.

  Таким образом, `sms_text` будет содержать текст SMS-сообщения в зависимости от условий и значений полей в выборке.   

7. Возвращает результат запроса, который представляет собой строки с данными о клиентах, подходящих под заданные условия.

Создадим результирующую таблицу с произвольным именем в текущей БД, записав все данные.

```python
# результирующую таблицу создадим с именем sms_base
cur.execute('''create table if not exists sms_base AS
        with temp_table(client_id, hash_id, rest_rub, tarif_name, tel_num, rank_cart, date_activated, neg)
         AS (
        select t.client_id, t.hash_id, u.rest_rub, r.tarif_name, i.tel_num,
        RANK() OVER(PARTITION BY t.client_id ORDER BY u.rest_rub desc) as rank_cart,
        t.date_activated, i.tel_num % 2
        from cards t

        left join tel_catalog i ON t.client_id=i.client_id
        left join accounts u ON t.acc_num=u.acc_num
        left join card_status y ON t.hash_id=y.hash_id
        left join catalog r ON t.tarif_id=r.tarif_id

        where t.client_id NOT IN (select * from blacklist x)
        and r.tarif_name = 'амурский тигр'
        and t.date_activated BETWEEN '2016-04-01' AND '2016-06-30'
        and u.rest_rub > 1000
        and DATE(u.report_date) = '2016-07-01'
        and y.status = 'open')


        select tt.client_id, tt.hash_id, tt.rest_rub, tt.tarif_name, tt.tel_num,
        case when tt.tel_num % 2 = 1
        then strftime('%m', tt.date_activated) ||', ' || (tt.rest_rub*1.0/1000) || ' тыс.руб, ' 
                 || substring(tt.tel_num, 2, 4)
        else 'Здравствуйте '|| tt.client_id || ', ' ||strftime('%m', tt.date_activated) ||', ' 
                 || (tt.rest_rub*1.0/1000) || ' тыс.руб, ' || substring(tt.tel_num, 2, 4) end as sms_text
        from temp_table tt
        where tt.rank_cart = 1


        ''')
```

```python
# проверим
sql = ''' select * from sms_base s'''
select(sql)
```
